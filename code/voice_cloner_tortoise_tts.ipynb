```python
import gradio as gr
import torch
import torchaudio
import tempfile
import numpy as np
from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio

tts = TextToSpeech()

def clone_voice(text, voice_sample_path):
    try:
        if not text or not voice_sample_path:
            return "Please provide both text and a voice sample."

        voice_sample = load_audio(voice_sample_path, 22050)

        gen = tts.tts_with_preset(text, voice_samples=[voice_sample], preset="fast")

        audio_array = gen.squeeze().cpu().numpy()

        audio_tensor = torch.from_numpy(audio_array).float().unsqueeze(0)

        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as temp_file:
            temp_filename = temp_file.name
            torchaudio.save(temp_filename, audio_tensor, 24000)

        return temp_filename
    except Exception as e:
        return str(e)

def process_request(text, voice_sample_path):
    output_audio = clone_voice(text, voice_sample_path)
    return output_audio

with gr.Blocks() as app:
    gr.Markdown("# Tortoise-TTS Voice Cloning")
    gr.Markdown("Upload a WAV file and enter text to generate speech.")

    with gr.Row():
        text_input = gr.Textbox(label="Enter Text")
        audio_input = gr.Audio(type="filepath", label="Upload Voice Sample")

    generate_button = gr.Button("Generate Speech")
    output_audio = gr.Audio(label="Generated Speech")

    generate_button.click(fn=process_request, inputs=[text_input, audio_input], outputs=output_audio)

app.launch(share=True, debug=True)
```
